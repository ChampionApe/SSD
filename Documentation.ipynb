# Documentation for CRRA-GHH model

This notebook shows the most important parts of the methods in ```CRRA.py```, albeit we do not cover anything. We cover the following:
1. Class structure.
2. The six main solution methods: $\lbrace$PEE, ESC, ESCB$\rbrace \times \lbrace$Time-dependent, steady state$\rbrace$.
3. Steady state methods.
4. Economic equilibrium methods.
5. Policy functions.
6. Terminal state solutions.
7. Reporting Functions

## 1. Class structure

The class structure is initialized with the ```__init__``` method, where we can adjust a number of settings already:

```python
    def __init__(self, ni = 11, T = 10, ngrid = 50, eps = 0.1, θ = .5, **kwargs)
```

From these inputs, the initialization method goes through a number of steps:
1. Add some of the information directly as attributes that can be accessed later (```self.ni, self.T, self.ngrid```).
2. Establish a main database, where parameters, variables, and indices/sets are generally stored. This is established at ```self.db``` which is a simple ```dict``` database. The database is initially populated with a range of default parameter values (see ```self.defaultParameters``` property), but these are replaced if other values are added using the ```**kwargs``` argument.
3. Define main indices/sets from the information on time, types, and gridsize: 
    * ```self.db['t']``` is a ```pd.RangeIndex``` from 0-T.
    * ```self.db['i']``` is ```pd.RangeIndex``` from 0-ni.
    * ```self.db['sgrid']``` is a ```pd.RangeIndex``` from 0-ngrid.
4. System characteristics ```eps``` and  ```θ``` are adjusted with special methods that makes sure that if they are entered as scalars, they are redefined over the time index as well, and the leaded versions ```eps[t+1]``` and  ```θ[t+1]``` are also added automatically to the ```self.db```.
5. An attribute called ```self.ns``` is added automatically. This is a dictionary with information on different types of numerical problems that are solved throughout the class. Each numerical problem is identified by a name (e.g. ```self.ns['EE']``` for the economic equilibrium) with information on what variables are solved in that problem (in this case ```s,h,Γs```). These "namespace" instances draws on a little homemade class ```SymMaps``` from the package ```pyDbs``` that are helpful when solving multivariate numerical functions (see https://github.com/ChampionApe/pyDbs/blob/main/SymMaps.ipynb for more).
6. An attribute called ```self.x0``` is added automatically. This is a dictionary that permits adding default initial values to various numerical problems. The keys in the dictionary corresponds to the ```solve``` methods of the class, and the values are default values that are used. For instance, the method ```self.solve_ss``` takes as input ```x0_ss = None```. When this is ```None```, the default initial value that is used is ```self.x0['ss']```.
7. An attribute called ```self.f0``` is added automatically. This is a dictionary that specifies which low-level functions higher-level functions are calling. For instance, the high-level method ```solve.steadyState_SA_PEE``` draws on either ```self.solve_PEE_t``` or the vectorized function ```self.solve_PEEvec_t```. This is specified and can be changed by adjusting ```self.f0['SA_PEE']``` to one or the other.

*Initialize model and adjust some parameter values:*

## 2. Four main solution methods identifying PEEs

The class contains a lot of auxiliary methods that are combined in different ways. The class contains four methods for identifying PEE paths: 
1. PEE, time-dependent: Politico-economic equilibrium with exogenous system characteristics ($\theta, \epsilon$), where the terminal period $T$ is solved as a steady state, and then time-dependent policy functions are solved for iterating backwards in time.
2. ESC, time-dependent: Similar to the first, except with Endogenous System Characteristics ($\theta_t,\epsilon_t$ are chosen in political equilibrium as well in each period).
3. PEE, steady state approx.: Identifies a steady state policy function on the two-dimensional grid $\tau(s_{-1}, \nu)$, and use this to approximate the PEE path over time. 
4. ESC, steady state approx.: Like 3., but with endogenous system characteristics as well.

In any case, before we can go to the solution, we need to set up a grid of states $s_{-1}$ to solve the model over (this is used in all four cases). Here, we define a nonlinear grid from (almost) zero to 125\% of the steady state level of savings when $\epsilon = \theta = \tau = 0$ (a combination that maximizes steady state savings):

````python
sGrid = CRRA.polGrid(1e-4, model.solve_ss(0,0,0,ν[-1])['s']*1.5, model.ngrid, 1.5) # nonlinear grid with more gridpoints in the lower part of the grid 
pd.Series(sGrid).plot(figsize=(4,3));
````

### 2.1. PEE, time-dependent 

```python
    def solve_PEE(self, s, gridOption = 'resample', s0 = None, returnPols = False, sskwargs = None, x0_t = False, tkwargs = None):
        ...
        return reportMain(sol)
``` 

The first main method is ```solve_PEE``` that takes five inputs: 
* The grid of states ```s``` (necessary)
* ```gridOption='resample'```: Specifies in what way the time-dependent policies are identified. Due to the endogenous gridpoint method, the gridpoints will naturally diverge over time if we do not handle this. That is, when we iterate backwards in time, the grid that starts on the interval $[0, \overline{s}]$ in period $t$ will lead to the interval $[0, (1+k)\overline{s}]$ in period $t-1$. As we are generally not interested in the policy function for values $>\overline{s}$, it makes sense to either resample the gridpoints within $[0, \overline{s}]$ in each period (```gridOption = 'resample'```) or truncate the solution and only use gridpoints within this interval (```gridOption = 'clean'```). If ```gridOption = None``` you get the full solution with diverging grids (not recommended).
* ```s0=None```: Once we have identified policy functions $\tau_t(s_{t-1}, \nu_t)$ for each $t$ and need to solve for the PEE path, we need to initialize the model with a level of savings $s_0$ that we enter the *first* period with. If ```s0 = None``` the model automatically computes the *steady state* level of savings that would occur in the first period $t$ and initialize the model with this (note that savings will not actually be in steady state because demographics and policy functions are changing over time). 
* ```sskwargs=None```: Allows you to pass kwargs to the steady state method used in ```self.solve_steadyState_PEE```. For more on these options, see section on this method below.
* ```tkwargs=None```: Allows you to pass kwargs to the out-of-steady state method used to identify policy functions (default method is ```self.solve_PEEvec_t```). For more on these options, see relevant section below.
* ```x0_t=False```: If set to ```True```, when looping through $t$, use initial values for the policy function that are specific to that year - i.e. make sure that ```self.x0['PEE_t']``` is a dictionary with initial values for each $t$. 

The method returns a dictionary with the main characteristics of the PEE path. This includes things like taxes, savings, labor supply etc.

**Notes:** 
* This method assumes that the system characteristics $\theta_t, \epsilon_t$ are defined as in the ```self.db```.
* The method relies on several other methods (some of which we return to below) that solves numerical problems. Whenever we rely on numerical optimziation, the code includes assertions that the solution is successful. Thus, if there is a numerical problem along the way, this returns an ```AssertionError``` in the relevant function.

*Default solution method, return policy functions as well:*

````python
sol_PEE, pol_PEE = model.solve_PEE(sGrid, returnPols = True)
````

## 3. Steady state methods

The class includes a handful of steady state methods that depends on whether or not inputs are scalars/vectors, choice variables are bounded or not, and whether or not to loop through vectors and solve or use vectorized functions. The main methods here are:
* ```self.solve_ss```: Solves for steady state levels of $B$, $\Gamma_s$, and $s$ given *scalar* inputs of $\tau, \theta, \epsilon,$ and $\nu$. Optimization solves for the vector $x = [\Gamma_s, B]$, where $B$ is a vector of values for each type $i$ of households ($s$ is computed afterwards). If initial values for $x$ is not given, the method relies on values from  ```self.x0['ss']```.
* ```self.solve_ssVec```: Solves for steady state lavels of $B$, $\Gamma_s$, and $s$ given *vector* of $\tau$; $\theta, \epsilon$, and $\nu$ are scalars. This method solves for the solution by looping through values of $\tau$ and solving problem akin to ```self.solve_ss```.
    * The length of the vectors should be ```self.ngrid```. 
    * If ```x0_ssVec = None``` numerical optimization uses default values from ```self.x0['ssVec']```.
    * If ```ite_ssVec = True``` the model loops through $\tau$ and uses initial values from the $n-1$ solution.
    * If ```ite_ssVec = False``` numerical optimization uses initial values from ```self.x0['ssVec']```. If we want to have new initial values for each loop, the initial values in ```self.x0['ssVec']``` needs to be a 2d array with number of rows corresponding to the lenght of the grid $\tau$.
* ```self.solve_ssVec_v2```: Similar to ```self.solve_ssVec```, but with vectorized numerical optimization. *NB: This is pretty slow, it does not handle sparse data very well.*
* ```self.solve_ss_ESC```: Similar to ```self.solve_ssVec```, but with $\theta, \epsilon$ being vectors of the same length as $\tau$. 

```self.solve_ss```:

````python
model.solve_ss(.15, .7, .1, 1, x0_ss = None) # τ = .15, θ = .7, ϵ = .1, ν = 1. Rely on default initial values for x0
````

```self.solve_ssVec```:

````python
τgrid = np.linspace(0, 1, model.ngrid)
ss = model.solve_ssVec(τgrid, .7, .1, 1, x0_ssVec = None, ite_ssVec = True) # default initial values for the first iteration - then use loop solutions
ss = model.solve_ssVec(τgrid, .7, .1, 1, x0_ssVec = np.hstack([ss['B'][:,0], ss['Γs'][0]]), ite_ssVec = True) # use custom initial value in the first iteration, then rely on loop
model.x0['ssVec'] = {i: np.hstack([ss['B'][:,i], ss['Γs'][i]]) for i in range(model.ngrid)} # Provide custom initial values 
ss = model.solve_ssVec(τgrid, .7, .1, 1, x0_ssVec = None, ite_ssVec = False) # use customized initial values for every iteration in the loop
pd.Series(ss['s']).plot(figsize = (4,3)); # plot solution of savings on the grid of taxes
````

## 4. Economic Equilibrium Functions

There are a lot of auxiliary functions that compute solutions for the economic equilibrium. We get to them in the section on "reporting functions". This section focuses on identifying the core variables of the economic equilibrium (that have to be solved numerically): $s_t, h_t$, and $\Gamma_{s,t}$ for each $t$. All methods rely on an initial value of savings ($s_0$) and assumes that the model terminates in a steady state solution in period $T$; thus the period $T$ equations all assumes $h_T = h_{T+1}, s_T = s_{T+1}$ etc..

There are three main solve methods for the economic equilibrium:
* ```self.solve_EE```: This solves for the economic equilibrium given vectors of $\tau, \theta, \epsilon$ (time-dimension) and initial value $s_0$. 
    * Numerical optimization solves for the stacked vector $x = [s, h, \Gamma_{s}]$. The linear index of the stacked variables can be accessed through the "namespace" ```self.ns['EE'].maps```. 
    * We can provide initial values using kwarg ```x0_EE``` or use default initial values (accessed from ```self.x0['EE']```) using ```x0_EE = None```.
    * The solution uses ```self.report_EE``` to add a couple of additional, important variables (currently, $s_{t-1}, \Theta_h, \Theta_s$).
* ```self.solve_EE_polFunc```: Solves the same system as ```self.solve_EE```, but where $\tau$ is no longer a vector of fixed policies, but a policy function (of $s_t, t$) that is constructed as a *gridded solution*. 
* ```self.solve_EE_ESC```: Similar to ```self.solve_EE_polFunc```, except that $\tau, \theta, \epsilon$ are all policy functions to be inferred from a gridded solution.

```self.solve_EE```: *Solution with constant policies and $s_0$ as the steady state value for $t=0$*

````python
τ, θ, ϵ = np.full(model.T, .2), np.full(model.T, .7), np.full(model.T, .1)
s0 = model.solve_ss(.2, .7, .1, model.db['ν'][0])['s'] # compute steady state value with fixed policy and worker-to-retiree ratio of the initial period
EE = model.solve_EE(τ, θ, ϵ, s0, x0_EE = None); # use initial values from self.x0['EE']
EE = model.solve_EE(τ, θ, ϵ, s0, x0_EE = np.hstack([EE[k].values for k in model.ns['EE'].symbols])) # solve with custom initial value, here taken directly from previous solution
EE['s'].plot(figsize=(4,3));
````

```self.solve_EE_polFunc```: *Solution with linearly increasing policies (identical for all $t$)*

````python
griddedSolution = {t: {'τ': np.linspace(0, 1, model.ngrid), 
                       's[t-1]': np.linspace(0,1,model.ngrid)} for t in model.db['t']} 
EE_polFunc = model.solve_EE_polFunc(griddedSolution, θ, ϵ, s0, x0_EE_polFunc = None); # use initial values from self.x0['EE_polFunc']
EE_polFunc = model.solve_EE_polFunc(griddedSolution, θ, ϵ, s0, x0_EE_polFunc = np.hstack([EE_polFunc[k].values for k in model.ns['EE'].symbols])) # solve with custom initial value, here taken directly from previous solution
EE_polFunc['s'].plot(figsize=(4,3));
````

```self.solve_EE_ESC```: *Solution with linearly increasing policies (identical for all $t$)*

````python
griddedSolution = {t: {'τ': np.linspace(0, 1, model.ngrid), 
                       'θ': np.linspace(0, 1, model.ngrid),
                       'eps': np.linspace(0, 1,model.ngrid),
                       's[t-1]': np.linspace(0,1,model.ngrid)} for t in model.db['t']} 
EE_ESC = model.solve_EE_ESC(griddedSolution, s0, x0_EE_ESC = None); # use initial values from self.x0['EE_polFunc']
EE_ESC = model.solve_EE_ESC(griddedSolution, s0, x0_EE_ESC = np.hstack([EE_ESC[k].values for k in model.ns['EE'].symbols])) # solve with custom initial value, here taken directly from previous solution
EE_ESC['s'].plot(figsize=(4,3));
````

## 5. Policy functions

We can identify policy functions in a number of different ways: First, we can either solve for $\tau$ only (PEE) or with endogenous system characteristics (ESC). Second, we can solve for *unbounded* or *bounded* solutions (ESCB). Third, we can solve using vectorized functions or looping through grids.

Before we showcase the policy functions, we start by reverting initial values to the default ones (we messed around with these in section 2):

### 5.1. ```self.solve_PEE_t``` and ```self.solve_PEEvec_t```

The two methods assume that $\theta, \epsilon$ are fixed, and that we have already solved a similar problem at time $t+1$ which - among other things - gives us a gridded policy function $\tau_{t+1}(s_t)$. To illustrate the two methods, we start by approximating this $t+1$ solution (```sol_p```) in steady state:

````python
θ, eps = .7, .1  # PEE values in t
θp, epsp = .7, .1 # PEE values in t+1
solPEE_p = model.steadyState_solp(np.full(model.ngrid, .2), θ, eps, ν[-1], sGrid)
````

The two methods solve for the same policy, but one does so using vectorized functions and one by looping. In this case, the vectorized function is somewhat faster ($\approx$ 2-4 times), but the loop one is more slightly more robust. 

````python
solPEE = model.solve_PEE_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p)
solPEE = model.solve_PEEvec_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p)
````

We can specify initial values in two ways:
* First, we can use ```ite_PEE_t = True``` (*default*): In this specification, the methods rely on the solution from the ```sol_p``` object (solution from t+1). For the loop-version of the model, this is only used in the first iteration, afterwards, the loop version rely on the solution from $n-1$ in the loop.
* Second, if ```ite_PEE_t = False```, we can specify initial values in ```self.x0['PEE_t']``` and ```self.x0['PEEvec_t']``` respectively.

#### Using ```ite_PEE_t = False```

Because the method ```solve_PEE_t``` is typically used when solving the model for each $t$, we can specify initial values in ```self.x0['PEE_t']``` in one of two ways. In the **first way**, we use keyword ```t=None``` to indicate that the initial values are *not* a function of time. This is the way the ```x0``` field is constructed per default:

````python
model.x0['PEE_t']
````

If we set ```ite_PEE_t = False``` (not default) and ```t=None``` (default), the methods rely on this numpy array for initial values. For the loop version, we extract the i'th value from the numpy array, for the vectorized version we use the entire array:

````python
model.getx0('PEE_t',t=None, i = 0), model.getx0('PEE_t',t=None) is model.x0['PEE_t']
````

This is called by adding:

````python
solPEE = model.solve_PEEvec_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p, ite_PEEvec_t = False)
solPEE = model.solve_PEE_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p, ite_PEE_t = False)
````

**The second way** that we can add initial values is to specify the kwarg $t$. Assume, for instance, that we wanted to add specific default values for the terminal period $t=T$. We do this by updating the initial values to a dictionary over time:

````python
model.x0['PEE_t'] = dict.fromkeys(model.db['t'], np.full(model.ngrid, .2)) # Use default guess for all t
model.x0['PEE_t'][model.T] = np.linspace(0.1,0.2, model.ngrid) # but adjust the guess for the terminal period.
model.x0['PEEvec_t'] = dict.fromkeys(model.db['t'], np.full(model.ngrid, .2)) # Use default guess for all t
model.x0['PEEvec_t'][model.T] = np.linspace(0.1,0.2, model.ngrid) # but adjust the guess for the terminal period.
````

We can specify that we should rely on these initial values as follows:

````python
solPEE = model.solve_PEEvec_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p, ite_PEEvec_t = False, t = model.T)
solPEE = model.solve_PEE_t(θ, eps, ν[-1], θp, epsp, ν[-1], solPEE_p, ite_PEE_t = False, t = model.T)
````

### 5.2. ```self.solve_ESC_t``` and ```self.solve_ESCvec_t```

The two methods are very similar to ```self.solve_PEE_t``` and ```self.solve_PEEvec_t```, except that they also endogenize $\theta, \epsilon$. As with the PEE case, we start with a steady state approximation to have something to start from:

````python
solESC_p = model.steadyState_solp_ESC(np.full(model.ngrid, .2), np.full(model.ngrid, .7), np.full(model.ngrid, .1), ν[-1], sGrid)
````

The two methods idenfity the same object, but one does so using scalar optimization and looping, the other vectorized. In this case, the vectorized version is roughly 1.5-2 times faster, but the loop version is more robust.

````python
solESC = model.solve_ESC_t(ν[-1], ν[-1], solESC_p)
solESC = model.solve_ESCvec_t(ν[-1], ν[-1], solESC_p)
````

We provide initial values in the same way as the corresponding PEE methods. If we set ```ite_ESC_t = False```, we default to using initial values from ```self.x0['ESC_t']``` instead:

````python
solESC = model.solve_ESCvec_t(ν[-1], ν[-1], solESC_p, ite_ESCvec_t = False) # default to initial values from numpy array from self.x0['ESC_t']
solESC = model.solve_ESC_t(ν[-1], ν[-1], solESC_p, ite_ESC_t = False)
````

We can also add that the initial values are different depending on $t$ (as with the PEE methods) by adding:

````python
model.x0['ESCvec_t'] = dict.fromkeys(model.db['t'], model.x0['ESCvec_t']) # repeat initial guess for all years 
# specify that for the terminal year, the initial guess are instead linearly spaced grids 
model.x0['ESCvec_t'][model.T] = np.hstack([np.linspace(0.1, 0.2, model.ngrid), # grid of taxes
                                           np.linspace(0.1, 0.3, model.ngrid), # grid of theta
                                           np.linspace(0.1, 0.3, model.ngrid)]) # grid of eps 
solESC = model.solve_ESCvec_t(ν[-1], ν[-1], solESC_p, ite_ESCvec_t = False, t = model.T) # use initial guess from t = terminal year T
````

Similarly, for the scalar-loop version:

````python
model.x0['ESC_t'] = dict.fromkeys(model.db['t'], model.x0['ESC_t']) # repeat initial guess for all years 
# specify that for the terminal year, the initial guess are instead linearly spaced grids 
model.x0['ESC_t'][model.T] = np.vstack([np.linspace(0.1, 0.2, model.ngrid), # grid of taxes
                                        np.linspace(0.1, 0.3, model.ngrid), # grid of theta
                                        np.linspace(0.1, 0.3, model.ngrid)]).T # grid of eps 
solESC = model.solve_ESC_t(ν[-1], ν[-1], solESC_p, ite_ESC_t = False, t = model.T) # use initial guess from t = terminal year T
````

### 5.3. ```self.solve_ESCB_t``` and ```self.solve_ESCBvec_t```

The two methods are equivalent to ```self.solve_ESC_t``` and ```self.solve_ESCvec_t```, except that the choices of $\theta, \epsilon$ are now bounded by [0,1]. The syntax is identical:

With the bounded version, the scalar-loop optimization seems to perform better both in terms of speed and numerical stability.

## 6. Terminal state solutions

We include three methods for "closing the model" by specifying terminal period functions:

1. Steady state assumption
2. Finite horizon assumption
3. Adhoc assumption.

### 6.1. Steady state

As in the previous section, we have steady state methods for the three types of solutions: PEE, ESC, and ESCB. In this section we outline the PEE case (the other two are simple extensions). The steady state solution is identified using a fixed-point requirement that the policy function $\tau(s_{-1})$ that comes from maximizing the political objective function *given* continuation policy $\tau_{+1}(s)$ coincides with the continuation policy on the entire grid.

To initialize the search for this fixedpoint, we can rely on ```self.steadyState_sol``` to solve for steady state *given* policy choices and demographics:

````python
τ, θ, eps = np.full(model.ngrid, 0.2), 0.7, 0.1 # initial assumptions for policies in the steady state; τ will be solved for endogenously.
ν_ss = ν[-1] # steady state demographics
sol_p = model.steadyState_solp(τ, θ, eps, ν_ss, sGrid)
````

Given this first guess, we have two ways of identifying the steady state policy: One is a successive approximation (SA) approach, the other is Newton-Kantorovich (NK). These methods rely on a couple of methods that are described in previous sections. 

#### 6.2. ```self.steadyState_SA_PEE```

The successive approximation approach ultimately relies on two steps that are repeated until convergence: Solving for the steady state (```steadyState_solp```) and identifying policy function given this (```self.solve_PEE_t```). As outlined above, we have different ways of identifying the two (loop-scalar vs. vectorization, initial values from iteration or default values).

**Identifying steady state solution:** As briefly discussed earlier, the loop-scalar version is generally both more robust and faster (due to scipy's poor performance with sparse data). Thus, this step always rely on the ```self.solve_ssVec``` method that uses loop-scalar optimization. In the first iteration, we can pass options to the ```self.steadyState_solp``` method using ```**kwargs```. In the next iterations, we always rely on the latest solution as initial guess.

````python
kwargs = {'x0_ssVec': np.array([0.1,0.15, 0.2, 0.25, 0.5])} # customize initial guess for the ssVec method in the first iteration
````

**Identifying solution given steady state:** As a default, the successive approximation uses the vectorized method ```self.solve_PEEvec_t``` to solve for this step. We can change this by updating ```self.f0['SA_PEE']``` though to a new function though. As with the steady state solution, this is called several times. So far, we only allow kwargs to be passed in the first iteration. This is passed as ```kwargs_SA0```.

````python
model.x0['PEEvec_t'] = np.linspace(0.1,0.2, model.ngrid) # specify customized initial values
kwargs_SA0 = {'ite_PEEvec_t': False} # add that the SA algorithm should rely on this in the first iteration instead
````

We can now identify the steady state policy by running (note that we need initial guess on taxes here as well):

````python
solPEE_ss = model.steadyState_SA_PEE(np.full(model.ngrid, 0.2), θ, eps, ν_ss, sGrid) # run with default options
solPEE_ss = model.steadyState_SA_PEE(np.full(model.ngrid, 0.2), θ, eps, ν_ss, sGrid, tol = 1e-6, iterMax = 100, kwargs_SA0 = kwargs_SA0, **kwargs) # identify steady state policy using SA approach with options
pd.Series(solPEE_ss, index = sGrid).plot(figsize=(4,3)); # plot steady state policy function
````

*Note: For the ```ESC``` and ```ESCB``` cases, the default option is loop-scalar optimization instead of vectorized functions; the reason is simply that these are more robust (which is more relevant here) and the speed gains are smaller for the vectorized version.*

````python
solESC_ss = model.steadyState_SA_ESC(np.hstack([np.full(model.ngrid, 0.2), 
                                                np.full(model.ngrid, 0.7), 
                                                np.full(model.ngrid, 0.1)]), ν_ss, sGrid) ## ESC with default options
model.x0['ESC_t'] = np.full((model.ngrid,3), .2) # re-arrange initial values for the loop version to work
solESC_ss = model.steadyState_SA_ESC(np.hstack([np.full(model.ngrid, 0.2), 
                                                np.full(model.ngrid, 0.7), 
                                                np.full(model.ngrid, 0.1)]), ν_ss, sGrid, tol = 1e-6, iterMax = 100, kwargs_SA0 = {'ite_ESC_t': False}, **{'x0_ss_ESC': np.array([0.1,0.15, 0.2, 0.25, 0.5])}) ## ESC with options
solESCB_ss = model.steadyState_SA_ESCB(np.hstack([np.full(model.ngrid, 0.2), 
                                                np.full(model.ngrid, 0.7), 
                                                np.full(model.ngrid, 0.1)]), ν_ss, sGrid) ## ESCB with default options
model.x0['ESCB_t'] = dict.fromkeys(range(model.ngrid), [0.1, 0.1, 0.1]) # Customize initial values
solESCB_ss = model.steadyState_SA_ESCB(np.hstack([np.full(model.ngrid, 0.2), 
                                                 np.full(model.ngrid, 0.7), 
                                                 np.full(model.ngrid, 0.1)]), ν_ss, sGrid, tol = 1e-6, iterMax = 100, kwargs_SA0 = {'ite_ESCB_t': False}, **{'x0_ss_ESC': np.array([0.1,0.15, 0.2, 0.25, 0.5])}) ## ESCB with options
pd.Series(model(solESCB_ss, 'τ'), index = sGrid).plot(figsize=(4,3)); # plot steady state policy function
````

#### 6.3. ```self.solve_steadyState_PEE```

This method is a mix of the Newton (NK) and the successive approximation (SA)approaches: 
* ```If tol_SA = None```: The steady state is identified using only NK method with tolerance ```tol_NK = 1e-6```.
* ```Elif tol_SA>tol_NK```: Steady state is identified first using SA, then NK method.
* ```Else:``` Steady state is identified using SA method.

This method always draw on initial guess on policy function from ```self.x0['steadyState_PEE']```. *Note: The SA algorithm has often proven faster and more robust than the NK algorithm here*.

We pass kwargs to the SA algorithm using ```kwargs_SA = None``` specified as a dictionary that is unpacked in the ```self.steadyState_SA_PEE``` function:

````python
solPEE, policyPEE = model.solve_steadyState_PEE(θ, eps, ν_ss, sGrid, tol_SA = 1e-6) # use SA algorithm only, default options 
policyPEE = model.solve_steadyState_PEE(θ, eps, ν_ss, sGrid, tol_SA = 1e-6, path = False, kwargs_SA = {'kwargs_SA0': kwargs_SA0}) # set path = False to return only policy grid, add kwargs_SA0 from earlier that is passed to SA solver
pd.Series(policyPEE['τ'], index = sGrid).plot(figsize = (4,3));
````

The NK algorithm only pass ```**kwargs```  directly to the scipy optimizer ```optimize.root```:

````python
solPEE, policyPEE = model.solve_steadyState_PEE(θ, eps, ν_ss, sGrid, tol_SA = None) # use NK algorithm only, default options
policyPEE = model.solve_steadyState_PEE(θ, eps, ν_ss, sGrid, tol_SA = None, path = False, **{'options': {'factor': .5}}) # adjust factor for initial step bound in 'hybr' algorithm
pd.Series(policyPEE['τ'], index = sGrid).plot(figsize=(4,3));
````

We can update the initial values by adjusting the ```self.x0``` field appropriately:

````python
model.x0['steadyState_PEE'] = policyPEE['τ']
policyPEE = model.solve_steadyState_PEE(θ, eps, ν_ss, sGrid, tol_SA = 1e-6, path = False) # this should make the solution a lot faster
````

*ESC syntax:*

````python
solESC, policyESC = model.solve_steadyState_ESC(ν_ss, sGrid, tol_SA = 1e-6) # use SA algorithm only, default options 
policyESC = model.solve_steadyState_ESC(ν_ss, sGrid, tol_SA = 1e-6, path = False, kwargs_SA = {'kwargs_SA0': {'ite_ESC_t': False}}) # set path = False to return only policy grid, tell that SA should not use iterative updates to solve
model.x0['steadyState_ESC'] = np.hstack([policyESC[k] for k in model.ns['ESC[t]'].symbols]) # update initial guess to make the NK part faster
policyESC = model.solve_steadyState_ESC(ν_ss, sGrid, tol_SA = None, path = False) # use NK algorithm only, default options
````

*ESCB syntax:*

````python
solESCB, policyESCB = model.solve_steadyState_ESCB(ν_ss, sGrid, tol_SA = 1e-6) # use SA algorithm only, default options 
policyESCB = model.solve_steadyState_ESCB(ν_ss, sGrid, tol_SA = 1e-6, path = False, kwargs_SA = {'kwargs_SA0': {'ite_ESCB_t': False}}) # set path = False to return only policy grid, tell that SA should not use iterative updates to solve
model.x0['steadyState_ESCB'] = np.hstack([policyESCB[k] for k in model.ns['ESC[t]'].symbols]) # update initial guess to make the NK part faster
policyESCB = model.solve_steadyState_ESCB(ν_ss, sGrid, tol_SA = None, path = False) # use NK algorithm only, default options
````
